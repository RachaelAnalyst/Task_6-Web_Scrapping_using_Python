```python
import requests
from bs4 import BeautifulSoup


def extract():
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }
    
    
    url = f"https://www.myjobmag.com/search/jobs?location=Rivers&currentpage=5"
    
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    
    return soup
```

```python
job_list_lagos=[]
```

```python
def transform (soup):
    
    lis= soup.find_all("li", class_="job-info")
    for job in lis:
        job_title= job.find("li", class_="mag-b").text.strip()
        job_date=job.find("li", class_="job-item").text.strip() + "-2026"
        job_location= "Lagos"
        job={
            "Title":job_title,
            "Date": job_date,
            "Location":job_location
        }
        job_list_lagos.append(job)
    return

c= extract()
transform(c)

```
